# E-Commerce Website Conversion A/B Testing Analysis 📈💡

## 1. Problem Statement & Business Context
An e-commerce company wants to test whether a newly designed product page (**Design B**) yields a higher conversion rate compared to the existing design (**Design A**). 

In digital marketing and product optimization, intuition can be costly. A/B testing allows businesses to use statistical inference to make data-driven decisions that directly affect user experience and revenue.

**Key Objectives:**
- Formulate and evaluate null and alternative hypotheses for conversion metrics.
- Simulate realistic e-commerce conversion data for control and treatment groups.
- Perform **Two-Sample T-Tests** for numerical continuous distributions.
- Perform **Chi-Square Tests of Independence** to analyze categorical conversion rates.
- Provide actionable business decisions based on statistical significance ($p$-values).

## 2. Setup & Environment Configurations

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import chi2_contingency
import warnings
warnings.filterwarnings('ignore')

# Set chart aesthetics
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

## 3. Data Simulation & Exploration
To model user interactions, we simulate two distinct user activities:
1. **Average Session Duration (Continuous Numerical):** Checking if users spend more time on the new layout.
2. **Conversion Counts (Categorical Discrete):** Checking if users actually complete a purchase (Convert vs. Non-Convert).

```python
# Seed for reproducibility
np.random.seed(42)

# 1. Simulate Session Durations (seconds) for 100 users in each group
duration_A = np.random.normal(loc=50, scale=10, size=100)
duration_B = np.random.normal(loc=54, scale=10, size=100)

print(f"Control Group (Design A) Mean Session Duration: {np.mean(duration_A):.2f} seconds")
print(f"Treatment Group (Design B) Mean Session Duration: {np.mean(duration_B):.2f} seconds")
```

### Visualizing Continuous Metric Distributions

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(duration_A, label="Control Group (Design A)", fill=True, color="#4C72B0", alpha=0.5)
sns.kdeplot(duration_B, label="Treatment Group (Design B)", fill=True, color="#55A868", alpha=0.5)
plt.axvline(np.mean(duration_A), color="#4C72B0", linestyle="--", linewidth=2)
plt.axvline(np.mean(duration_B), color="#55A868", linestyle="--", linewidth=2)

plt.title("User Session Duration Distribution (Design A vs. Design B)", fontsize=14)
plt.xlabel("Session Duration (seconds)")
plt.ylabel("Density")
plt.legend(loc="upper right")
plt.show()
```

## 4. Hypothesis Testing (Continuous Metric: Session Duration)

We want to determine if the variance/mean difference is statistically significant using a **Two-Sample Independent T-Test**.

- **Null Hypothesis ($H_0$):** There is no significant difference in the mean session duration between Design A and Design B ($\mu_A = \mu_B$).
- **Alternative Hypothesis ($H_1$):** There is a statistically significant difference in the mean session duration between Design A and Design B ($\mu_A \neq \mu_B$).

```python
# Execute Independent Two-Sample T-Test
t_stat, p_value_t = stats.ttest_ind(duration_A, duration_B)

print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value_t:.6f}")

alpha = 0.05
if p_value_t < alpha:
    print("\nDecision: Reject the Null Hypothesis (H0).")
    print("Conclusion: The difference in session duration between the two groups is statistically significant.")
else:
    print("\nDecision: Fail to reject the Null Hypothesis (H0).")
    print("Conclusion: There is no statistically significant evidence that session durations differ.")
```

## 5. Hypothesis Testing (Categorical Metric: Conversion Rates)

Now let's examine user conversions (whether they bought an item or not). We set up a contingency table showing the actual conversion numbers for both groups:
- **Design A (Control):** 30 converted, 70 did not convert (30% Conversion Rate)
- **Design B (Treatment):** 50 converted, 50 did not convert (50% Conversion Rate)

We run a **Chi-Square ($\chi^2$) Test of Independence** to see if the version of the layout impacts conversion behavior.

- **Null Hypothesis ($H_0$):** Layout design and conversion behavior are independent (the new design has no true impact on conversion rate).
- **Alternative Hypothesis ($H_1$):** Layout design and conversion behavior are dependent (the new design significantly impacts conversion rate).

```python
# Create Contingency Table [[Converted, Not Converted]]
contingency_table = [
    [30, 70],  # Design A
    [50, 50]   # Design B
]

# Perform Chi-Square Test
chi2_stat, p_value_chi2, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"P-Value: {p_value_chi2:.6f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies Table:\n", expected)

if p_value_chi2 < alpha:
    print("\nDecision: Reject the Null Hypothesis (H0).")
    print("Conclusion: Conversion rates are significantly dependent on the webpage layout design.")
else:
    print("\nDecision: Fail to reject the Null Hypothesis (H0).")
    print("Conclusion: No statistically significant difference detected in conversion rates.")
```

### Visualizing Conversion Rate Differences

```python
conversion_data = pd.DataFrame({
    'Group': ['Control (Design A)', 'Treatment (Design B)'],
    'Conversion Rate (%)': [30.0, 50.0]
})

plt.figure(figsize=(6, 5))
ax = sns.barplot(x='Group', y='Conversion Rate (%)', data=conversion_data, palette='Set1')
plt.title('Conversion Rate Comparison (A/B Test)', fontsize=14)
plt.ylim(0, 100)

# Add data labels
for p in ax.patches:
    ax.annotate(f"{p.get_height():.1f}%", (p.get_x() + p.get_width() / 2., p.get_height() + 2),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points', weight='bold')

plt.ylabel('Conversion Rate (%)')
plt.show()
```

## 6. Comprehensive Summary & Executive Insights 🏁

### Summary of Analytical Findings:
1. **Continuous Metric (Session Duration):** The independent t-test yielded a statistically significant result ($p < 0.05$), meaning users spend noticeably more time engaging with **Design B** compared to **Design A**.
2. **Categorical Metric (Conversion Rate):** The Chi-Square test confirmed that the jump from a **30% conversion rate to a 50% conversion rate** is highly significant statistically ($p$-value is well below the 0.05 threshold).

### Strategic Business Recommendations:
- **Rollout Strategy:** Since both continuous engagement and absolute business conversion metrics show highly statistically significant improvement, **Design B should be rolled out fully** to replace Design A.
- **ROI Estimation:** Shifting from a 30% baseline to a 50% conversion represents a **66.7% relative surge in transactions**, which directly correlates with a substantial scale-up in sales revenue without any added customer acquisition costs.
